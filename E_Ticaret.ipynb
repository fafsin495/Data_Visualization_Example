```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\fafsi\\OneDrive\\Masaüstü\\DATAITEam\\Yapay Zeka İçin Python Programlama(1)\\E_ticaret2015-2020.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
df.drop("UNIXTIME",axis=1,inplace=True)
df.drop("Tarih",axis=1,inplace=True)
df.tail()
```

```python
df["tl"] = df["TP_KKHARTUT_KT50"]

df["tarih"] = df["YEARWEEK"]
df.drop("TP_KKHARTUT_KT50",axis=1,inplace=True)
df.drop("YEARWEEK",axis=1,inplace=True)
```

```python
onbes = df[:52]
onalti =  df[52:105]
onyedi= df[105:157]
onsekiz = df[157:209]
ondokuz= df[209:261]
yirmi = df[261:279]
```

# 2015

```python


#grafiğimizin boyutu
plt.figure(figsize=(16,8))

#çizimi
onbes
sns.set_style("whitegrid")
p1 = sns.pointplot(x=onbes.tarih,y=onbes.tl,color="blue",alpha=0.5)

#grafik üzerine değerlerin yazılması

for i in range (0,onbes.shape[0]):
    p1.text(i,onbes.tl.iloc[i]+0.03,onbes.tl.iloc[i],size="medium",color="red",weight="semibold")
    
#x eksenine yazıları 90 derece açıyla yazma 
plt.xticks(rotation=90)

#x ve y eksenimizi isimlendirdik 

plt.xlabel("Tarih",fontsize=15)
plt.ylabel("Bin-Tl",fontsize=15)
plt.title("E-Ticaret Oranı-2015",fontsize=20)

#oluşturduğumuz grafiği kaydettik
picturename = "E-Ticaret Oranı-2015.png".format(onbes.tl,onbes.iloc[onbes.shape[0]-1].name)
plt.savefig(picturename , dpi=200)
plt.show()
```
